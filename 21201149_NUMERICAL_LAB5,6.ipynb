# **Question 01 :**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return 4*x**5 - 9*x**2 + 6*x - 10

def f_prime(x):
    return 20*x**4 - 18*x + 6
```

```python
true_value = f_prime(3)
```

**FDD:**

```python
def fdd(f, x, h):
    return (f(x + h) - f(x)) / h
```

**BDD:**

```python
def bdd(f, x, h):
    return (f(x) - f(x - h)) / h
```

**CDD:**

```python
def cdd(f, x, h):
    return (f(x + h) - f(x - h)) / (2 * h)
```

**Approximate Errors for FDD:**

```python
prev_approx = fdd(f, 3, 0.3)
print("First Approximation: ", prev_approx)
print()

h = [0.15, 0.1, 0.05, 0.01, 0.001, 0.0001]
for i in h:
    present_approx = fdd(f, x=3, h=i)

    print("Previous approximation for h = ", i, " is: ", prev_approx)
    print("Present approximation for h = ", i, " is: ", present_approx)

    AE = abs(present_approx - prev_approx)

    print("AE: ", AE)
    print()
    prev_approx = present_approx
```

**Approximate Errors for BDD:**

```python
prev_approx = bdd(f, 3, 0.3)
print("First Approximation: ", prev_approx)
print()

h = [0.15, 0.1, 0.05, 0.01, 0.001, 0.0001]
for i in h:
    present_approx = bdd(f, x=3, h=i)

    print("Previous approximation for h = ", i, " is: ", prev_approx)
    print("Present approximation for h = ", i, " is: ", present_approx)

    AE = abs(present_approx - prev_approx)

    print("AE: ", AE)
    print()
    prev_approx = present_approx
```

**Approximate Errors for CDD:**

```python
prev_approx = cdd(f, 3, 0.3)
print("First Approximation: ", prev_approx)
print()

h = [0.15, 0.1, 0.05, 0.01, 0.001, 0.0001]
for i in h:
    present_approx = cdd(f, x=3, h=i)

    print("Previous approximation for h = ", i, " is: ", prev_approx)
    print("Present approximation for h = ", i, " is: ", present_approx)

    AE = abs(present_approx - prev_approx)

    print("AE: ", AE)
    print()
    prev_approx = present_approx
```

**Plot :**

```python
prev_approx1 = fdd(f, x=3, h=0.3)
prev_approx2 = bdd(f, x=3, h=0.3)
prev_approx3 = cdd(f, x=3, h=0.3)

h = [0.15, 0.1, 0.05, 0.01, 0.001, 0.0001]
fdd_errors = []
bdd_errors = []
cdd_errors = []

for i in h:
    present_approx = fdd(f, x=3, h=i)
    AE = abs(present_approx - prev_approx)
    fdd_errors.append(AE)
    prev_approx = present_approx

print("H Values: ", h)
print("Errors: ", fdd_errors)

for i in h:
    present_approx = bdd(f, x=3, h=i)
    AE = abs(present_approx - prev_approx)
    bdd_errors.append(AE)
    prev_approx = present_approx

print("H Values: ", h)
print("Errors: ", bdd_errors)

for i in h:
    present_approx = cdd(f, x=3, h=i)
    AE = abs(present_approx - prev_approx)
    cdd_errors.append(AE)
    prev_approx = present_approx

print("H Values: ", h)
print("Errors: ", cdd_errors)

plt.figure(figsize=(10, 5))
plt.plot(h, fdd_errors, label='FDD', marker='o')
plt.plot(h, bdd_errors, label='BDD', marker='^')
plt.plot(h, cdd_errors, label='CDD', marker='x')
plt.xlabel("h", fontdict={'fontsize': 15})
plt.ylabel("AE", fontdict={'fontsize': 15})
plt.title("Approximate Error vs Step Size", fontdict={'fontsize': 20})
plt.show()
```

```python
x = np.linspace(-1, 1, 1000)
h = 0.05

der_fdd = fdd(f, x, h)
der_bdd = bdd(f, x, h)
der_cdd = cdd(f, x, h)
der_true = f_prime(x)

plt.figure(figsize=(10, 5))
plt.plot(x, der_fdd, label='FDD')
plt.plot(x, der_bdd, label='BDD')
plt.plot(x, der_cdd, label='CDD')
plt.plot(x, der_true, label='True')
plt.xlabel('$x$', fontsize=15)
plt.ylabel('$f\'(x)$', fontsize=15)
plt.title('Comparison of Differentiation Methods', fontsize=20)
plt.legend()
plt.show()
```
